```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
```

```python
iris = datasets.load_iris()
```

```python
X = pd.DataFrame(iris.data)
```

```python
X.head()
```

```python
X.columns = ['s_length', 's_width','p_length','p_width']
```

```python
X.head()
```

```python
a = KMeans(n_clusters = 2, init = 'k-means++')

y = a.fit_predict(X)
```

```python
a.labels_
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
colormap = np.array(['Red','Blue'])
Z = plt.scatter(X.s_length,X.s_width, c = colormap[a.labels_])
```

```python

plt.scatter(a.cluster_centers_[:,0], a.cluster_centers_[:,1], s= 150, c = 'Black', label = 'Centroids')
```

```python
colormap = np.array(['Red','Blue'])
plt.scatter(X.s_length,X.s_width, c = colormap[a.labels_])
plt.scatter(a.cluster_centers_[:,0], a.cluster_centers_[:,1], s= 150, c = 'Black', label = 'Centroids')
```

```python
accuracy_score(iris.target, y)
```

Elbow Method

```python
Error =[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(X)
    Error.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 11), Error)
plt.title('Elbow-Method using WCSS with k=1-11')
plt.xlabel('No of clusters')
plt.ylabel('Error') #within cluster sum of squares
plt.show()
```

No of Cluster = 3

```python
a = KMeans(n_clusters = 3, init = 'k-means++')

y = a.fit_predict(X)
```

```python
colormap = np.array(['Red','Blue','Green'])
plt.scatter(X.s_length,X.s_width, c = colormap[a.labels_])
plt.scatter(a.cluster_centers_[:,0], a.cluster_centers_[:,1], s= 150, c = 'Black', label = 'Centroids')
```

```python
accuracy_score(iris.target, y)
```
